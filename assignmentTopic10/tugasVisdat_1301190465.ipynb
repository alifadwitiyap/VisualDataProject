#### Assignment
- Buatlah visualisasi jaringan yang merepresentasikan jalur transportasi yang menghubungkan kota-kota besar di Jawa Barat.

- Tentukan kota terpenting berdasarkan nilai degree centrality dan betweenness centrality.

#### Answer

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
kota=[
"Bogor",
"Sukabumi",
"Cianjur",
"Bandung",
"Garut",
"Tasikmalaya",
"Ciamis",
"Kuningan",
"Cirebon",
"Majalengka",
"Sumedang",
"Indramayu",
"Subang",
"Purwakarta",
"Karawang",
"Bekasi",
"Pangandaran",
"Bogor",
"Sukabumi",
"Cirebon",
"Bekasi",
"Depok",
"Cimahi",
"Tasikmalaya",
"Banjar",
]
```

```python
G = nx.Graph()
G.add_nodes_from(kota)
```

```python
print(sorted(G.nodes()))
```

```python
G.add_edge('Depok','Bekasi')
G.add_edge('Depok','Bogor')
G.add_edge('Bekasi','Karawang')
G.add_edge('Bekasi','Karawang')
G.add_edge('Bogor','Sukabumi')
G.add_edge('Bogor','Cianjur')
G.add_edge('Bogor','Karawang')
G.add_edge('Sukabumi','Cianjur')
G.add_edge('Karawang','Subang')
G.add_edge('Karawang','Purwakarta')
G.add_edge('Karawang','Cianjur')
G.add_edge('Cianjur','Garut')
G.add_edge('Cianjur','Bandung')
G.add_edge('Cianjur','Cimahi')
G.add_edge('Cianjur','Purwakarta')
G.add_edge('Purwakarta','Cimahi')
G.add_edge('Purwakarta','Bandung')
G.add_edge('Purwakarta','Subang')
G.add_edge('Subang','Indramayu')
G.add_edge('Subang','Sumedang')
G.add_edge('Subang','Cimahi')
G.add_edge('Subang','Bandung')
G.add_edge('Indramayu','Cirebon')
G.add_edge('Indramayu','Majalengka')
G.add_edge('Indramayu','Sumedang')
G.add_edge('Sumedang','Cimahi')
G.add_edge('Sumedang','Bandung')
G.add_edge('Sumedang','Garut')
G.add_edge('Sumedang','Majalengka')
G.add_edge('Garut','Bandung')
G.add_edge('Garut','Cimahi')
G.add_edge('Garut','Tasikmalaya')
G.add_edge('Cimahi','Bandung')
G.add_edge('Tasikmalaya','Ciamis')
G.add_edge('Ciamis','Kuningan')
G.add_edge('Ciamis','Majalengka')
G.add_edge('Majalengka','Kuningan')
G.add_edge('Majalengka','Cirebon')
G.add_edge('Kuningan','Cirebon')
G.add_edge('Banjar','Ciamis')
G.add_edge('Banjar','Kuningan')
G.add_edge('Banjar','Tasikmalaya')
G.add_edge('Pangandaran','Banjar')
G.add_edge('Pangandaran','Ciamis')
G.add_edge('Pangandaran','Tasikmalaya')
```

```python
print(sorted(G.edges))
```

```python
plt.figure(figsize=(20,12))
pos = nx.spring_layout(G)

nx.draw(G,pos,with_labels = True, node_size=10,font_size=20)
```

```python
Centrality=nx.degree_centrality(G)

sort_orders = sorted(Centrality.items(), key=lambda x: x[1], reverse=True)

print(f'kota terpenting berdasarkan degree centrality\n')
for i in sort_orders:
	print(i[0], i[1])
```

```python
plt.figure(figsize=(20,12))
pos = nx.spring_layout(G)

nx.draw(G,pos, node_size=10)

for node, (x, y) in pos.items():
    plt.text(x, y, node, fontsize=Centrality[node]*100, ha='center', va='center')
```

```python

betweeness=nx.betweenness_centrality(G)

sort_orders = sorted(betweeness.items(), key=lambda x: x[1], reverse=True)

print(f'kota terpenting berdasarkan betweenness centrality\n')
for i in sort_orders:
	print(i[0], i[1])
```

```python
plt.figure(figsize=(20,12))
pos = nx.spring_layout(G)

nx.draw(G,pos, node_size=10)

for node, (x, y) in pos.items():
    plt.text(x, y, node, fontsize=betweeness[node]*100+10, ha='center', va='center')
```
