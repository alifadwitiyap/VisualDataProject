```python
df = pd.read_csv('https://raw.githubusercontent.com/alifadwitiyap/dataHosting/master/Indonesia_coronavirus_daily_data.csv')
df
```

```python
dfCov1Mei = df.loc[df["Date"] == "2021-05-01"]
dfCov1Mei.sort_values(by=['Daily_Case'],ascending=False)
```

```python
dfMerge = map.merge(dfCov1Mei, left_on="PROVINSI", right_on="Province")
dfMerge.head()
```

```python
ax = gplt.choropleth(
    dfMerge,
    hue="Daily_Case",
    edgecolor="white",
    linewidth=2,
    legend=True,
    cmap="Reds",
    scheme="FisherJenks",
    alpha = 1,
    projection=gcrs.AlbersEqualArea(),
    figsize=(20, 20)
)

gplt.polyplot(
    map,
    ax=ax,
    edgecolor="black",
    linewidth=2,
    zorder=2
)

ax.set_title(
    "Peta Sebaran Kasus Positif Harian COVID19 di Indonesia per Tanggal 1 Mei 2021 Basis Provinsi",
    fontsize=18
)
```
