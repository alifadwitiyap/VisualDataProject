```python
# import library yang diperlukan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap
from wordcloud import WordCloud
from PIL import Image
from imageio import imread
from collections import Counter
from matplotlib.animation import FuncAnimation
%matplotlib inline

# color pallet Netflix
pallet = ['#f5f5f1', '#e50914', '#b20710', '#221f1f']
cmapNetflix = LinearSegmentedColormap.from_list(name="netflix", colors=pallet)
sns.palplot(pallet)
```

#### Clean data


```python
df = pd.read_csv(
    'https://raw.githubusercontent.com/alifadwitiyap/VisualDataProject/master/assignmentTopic8/data/netflix_titles.csv')
df.head(5)
```

```python
# Missing data

def checkMissing(df):
    print("null value: ")
    for col in df.columns:
        null = round(df[col].isna().sum() / len(df) * 100, 2)
        if null > 0:
            print(f"{col} null rate: {null}%")


print("sebelum dibersihkan\n")
checkMissing(df)
print(f"shape : {df.shape}")


# isi dengan unknown
df = df.fillna("unknown")

# drop duplicate
df.drop_duplicates(inplace=True)

print("\n \nsetelah dibersihkan\n")
checkMissing(df)
print(f"shape : {df.shape}")
```

#### data visualization


```python
data = df['type'].value_counts().reset_index().values.T
data[1] = data[1]/data[1].sum()*100
labels = data.T
labels = [f"{l[0]} \n ({l[1]:.1f}%)" for l in labels]

plt.figure(figsize=(12, 7))
plt.pie(data[1],
        labels=labels,
        radius=1,
        labeldistance=1.2,
        colors=pallet[2:4],
        shadow=True,
        textprops={"weight": "bold"},
        wedgeprops={"edgecolor": "white",
                    'linewidth': 0.2},
        startangle=60
        )

plt.title("Presentase Tipe Film di Netflix",
          fontsize=16,
          fontweight='bold'
          )


# add a circle at the center to transform it in a donut chart
my_circle = plt.Circle((0, 0), 0.5, color='white')
p = plt.gcf()
p.gca().add_artist(my_circle)

plt.text(
        -0.1,-0.1,f'Total\n{df.shape[0]}'
         ,fontdict={
                 'weight': 'bold',
                 'size':14}
         )

plt.show()
```

```python
df['year'] = df['date_added'].apply(lambda x: x.split(', ')[-1])
df['month'] = df['date_added'].apply(lambda x: x.lstrip().split(' ')[0])

month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December'][::-1]

plt.figure(figsize=(10, 7))
data = df.groupby('year')['month'].value_counts(
).unstack().fillna(0)[month_order].T.iloc[:, :-1]

plt.pcolor(data, cmap=cmapNetflix, edgecolors='white', linewidths=2)

plt.xticks(np.arange(0.5, len(data.columns), 1), data.columns, fontsize=10)
plt.yticks(np.arange(0.5, len(data.index), 1), data.index, fontsize=10)

plt.title('Update Konten di Netflix',
          fontsize=16,
          fontweight='bold')
plt.colorbar()

plt.show()
```

```python
data = df.country.value_counts()
data = data.drop('unknown')[:10]


plt.figure(figsize=(12, 7))
colorMapper = [pallet[0] for _ in range(10)]
colorMapper[0] = colorMapper[1] = colorMapper[2] = pallet[2]

plt.bar(data.index, data, width=0.5,
        edgecolor='darkgray',
        linewidth=0.6, color=colorMapper)

for i in data.index:
    plt.annotate(f"{data[i]}",
                 xy=(i, data[i] + 150),
                 va='center', ha='center', fontweight='light', fontfamily='serif')

plt.title("Negara Produksi Film Netflix Terbanyak",fontdict={'fontsize':16,'fontweight':'bold'})
plt.grid(axis='y', linestyle='-', alpha=0.4)
plt.yticks(np.arange(0, 4000, 500))

plt.show()

```

```python
data = df.rating.value_counts()[:10]

plt.figure(figsize=(12, 7))
colorMapper = [pallet[0] for _ in range(10)]
colorMapper[0] = colorMapper[1] = colorMapper[2] = pallet[2]

plt.bar(data.index, data, width=0.5,
        edgecolor='darkgray',
        linewidth=0.6, color=colorMapper)

for i in data.index:
    plt.annotate(f"{data[i]}",
                 xy=(i, data[i] + 150),
                 va='center', ha='center', fontweight='light', fontfamily='serif')

plt.title("Rating Film Netflix Terbanyak",fontdict={'fontsize':16,'fontweight':'bold'})
plt.grid(axis='y', linestyle='-', alpha=0.4)
plt.yticks(np.arange(0, 4000, 500))

plt.show()
```

```python
# make input
#@title{ run: "auto" }
year="all time" #@param ['all time',2008, 2009, 2010, 2011, 2012, 2013,2014,2015,2016,2017,2018,2019,2020,2021]

if year == "all time":
  data=df.loc[:,:]
else:
  data=df.loc[df.year==year,:]

# create dictionary
combinedString=""

for e in data['listed_in']:
  combinedString+=e+', '

word=combinedString.split(', ')
word=dict(Counter(word))

# plot wordcloud
plt.figure(figsize=(12, 7))
wordcloud = WordCloud(background_color = 'white', width = 500,  height = 200,colormap=cmapNetflix,relative_scaling=0,max_words=15).generate_from_frequencies(word)
plt.imshow(wordcloud, interpolation="bilinear")
plt.title('Netflix Genre Wordcloud in '+year+'\n',
          fontsize=16,
          fontweight='bold')
plt.axis('off')
plt.show()
```
